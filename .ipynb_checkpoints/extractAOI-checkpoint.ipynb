# Usability Analysis Report

## Executive Summary

This report presents a comprehensive analysis of usability studies conducted on three travel booking websites: Booking.com, Expedia.com, and Trip.com. The analysis integrates quantitative metrics from eye-tracking and click-tracking, qualitative insights from think-aloud protocols, and patterns from browser logs to identify usability issues and recommend improvements.

## Methodology

The analysis followed a structured approach, examining time to first fixation, fixation duration, number of fixations, saccade to fixation ratio, number of saccades, saccade amplitude, scan path length, spatial density of scan path, total fixation time, ratio of eye path to task length, and mean pupil size. Qualitative data from think-aloud protocols and browser logs were also reviewed to contextualize the quantitative findings.

## Findings

### Booking.com

- **Time to First Fixation:** Longer than competitors, suggesting initial difficulty in locating a starting point, potentially due to layout complexity or unclear visual cues.
- **Total Fixation Time:** Higher, indicating extensive information processing time, likely due to pop-up interruptions and confusing navigation.
- **Spatial Density of Scan Path:** Lower, implying a more extensive search across the page, possibly due to visual clutter or ineffective information architecture.

**Qualitative Insights:**
- Users reported frequent pop-up interruptions and intrusive sign-up prompts, which align with the longer time to first fixation.
- Confusing navigation and UI elements were noted, supported by the high total fixation time and extensive scan paths.
- Issues with filtering and sorting were evident, correlating with the lower spatial density of scan paths.

**Browser Logs:**
- Rapid clicks and page reloads suggest confusion or lack of feedback, consistent with reported pop-up interruptions and navigation issues.

### Expedia.com

- **Saccade to Fixation Ratio:** Higher, indicating more scanning relative to reading, which may reflect unclear content or difficulty locating relevant information.
- **Total Fixation Time:** Lower than Booking.com, suggesting less engagement or usefulness of content.

**Qualitative Insights:**
- Users experienced confusion with site terminology and navigation labels, which is consistent with the higher saccade to fixation ratio.
- Inconsistent and misleading pricing was a significant source of frustration, aligning with rapid clicks observed in browser logs.

**Browser Logs:**
- Repeated clicking on price change notifications and rapid scrolling indicate confusion and difficulty in finding budget-appropriate options.

### Trip.com

- **Scan Path Length:** Longest among the three, suggesting inefficient layout or confusing content structure.
- **Spatial Density of Scan Path:** Lowest, indicating a more widespread search across the page, possibly due to a lack of clear visual hierarchy.

**Qualitative Insights:**
- Users struggled with the calendar interface and applying filters, which likely contributed to the longer scan path length and lower spatial density.
- Inconsistent UI components and system behavior were reported, supporting the observed extensive search behavior.

**Browser Logs:**
- Rapid clicks on the date picker and unexpected page reloads suggest technical issues or user error, consistent with reported confusion with date selection and system behavior.

## Recommendations

### Booking.com

- Minimize pop-ups and streamline sign-up processes to reduce initial fixation time.
- Clarify navigation and UI elements to decrease total fixation time and scan path length.
- Enhance filter functionality and persistence to improve spatial density of scan paths.

### Expedia.com

- Clarify terminology and navigation labels to reduce scanning behavior.
- Ensure transparent and consistent pricing to minimize user frustration and rapid clicks.

### Trip.com

- Improve the date picker interface to reduce scan path length.
- Clarify filter functionality and provide better feedback to enhance spatial density of scan paths.

## Conclusion

The analysis revealed distinct usability issues across the three websites, with common themes around navigation clarity, filter functionality, and information presentation. Addressing these issues through the recommended improvements should lead to a more intuitive, efficient, and satisfying user experience.
